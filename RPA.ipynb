```
import numpy as np
import pandas as pd
import pyautogui
import pyperclip
import subprocess
import sys
from pandas import Series,DataFrame
from time import sleep

secs = 1
pyautogui.PAUSE = 0.05
pyautogui.FAILSAFE = True

def pyperpaste():
    pyautogui.keyDown("ctrl")
    pyautogui.press("v")
    pyautogui.keyUp("ctrl")
    
    
#現在開いているwindowの数
opening_windows = subprocess.run(["wmctrl","-l"],stdout=subprocess.PIPE).stdout.decode('utf-8')
opening_windows = [i for i in opening_windows.split("\n") if i != ""]
opening_windows = int(len(opening_windows)) - 2


#全画面下部に収納
pyautogui.hotkey('ctrl','alt','d')
#windowを一つ選択
pyautogui.keyDown("alt")
for _ in range(2):
    pyautogui.press("tab")
pyautogui.keyUp("alt")
#左上に移動
pyautogui.hotkey('alt','l')


#Excelデータ読み込み
random_dammy = pd.read_excel('autotest.xlsx', sheetname='Sheet1')
#読み込んだdataframeをnumpy形式のmatrixに変換
dammy_data = DataFrame.as_matrix(random_dammy,columns=None)


#実行確認
choice = input("開始しますか？ (y/n)").lower()
if choice == "y":
    pyautogui.moveTo(86, 253, duration=0.2)
    pyautogui.click()

    first_time_wait = True
    for j in dammy_data:
        for i in j:
            pyperclip.copy(str(i))
            if first_time_wait:
                sleep(secs)
                first_time_wait = False
            pyautogui.press("space")
            pyautogui.press("backspace")
            pyperpaste()
            #隣のセルに移動する為のtab
            pyautogui.press("tab")
        #改行の為のtab
        pyautogui.press("tab")
else:
    sys.exit()
```
